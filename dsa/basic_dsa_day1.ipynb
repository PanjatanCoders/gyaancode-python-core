## Q1. Find the Maximum Number in an Array.
```
Example:
Input: [1, 4, 2, 7, 3]
Output: 7
```
Hint:

Walk through the list and keep track of the biggest thing you’ve seen so far.

````python
arr = [2, 4, 3, 6, 4, 49, 3, 24]
max = arr[0]

for num in arr:
    if max < num:
        max = num

print(f"Maximum nuber in arr {arr} is {max}")
````

## Q2: Reverse an Array (without using reverse())
```
Example:
Input: [1, 2, 3, 4]
Output: [4, 3, 2, 1]
Hint:
```
Swap left and right until they meet.

````python
arr = [1, 2, 3, 4]
left_index = 0
right_index = len(arr) - 1

while(left_index < right_index):
    arr[left_index], arr[right_index] = arr[right_index], arr[left_index]
    left_index +=1
    right_index -=1
    
print(f'After reversing the arr is: {arr}')
````

## ## :Q4: Move All Zeroes to the End (keeping order)
```
Example:
Input: [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]
```
Hint:

Use a pointer to track the next non-zero write position.

````python
nums = [0, 1, 0, 3, 12]

j = 0

for i in range(len(nums)):
    if nums[i] != 0:
        nums[i], nums[j] = nums[j], nums[i]
        j += 1

print(f'Final output: {nums}')
````
